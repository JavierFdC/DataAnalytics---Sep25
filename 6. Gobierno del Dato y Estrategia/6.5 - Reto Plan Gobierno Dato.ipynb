## Fase 2 — Matriz RACI de Procesos de Gobierno (10 min)

La matriz RACI asigna responsabilidades claras a cada rol para cada proceso de gobierno. Recordemos los cuatro niveles:

| Letra | Significado | ¿Quién? |
|-------|------------|---------|
| **R** | Responsible (ejecuta) | Quien hace el trabajo operativo |
| **A** | Accountable (rinde cuentas) | Quien toma la decisión final — solo uno por proceso |
| **C** | Consulted (consultado) | Quien aporta conocimiento antes de decidir |
| **I** | Informed (informado) | Quien recibe notificación después de la acción |

En la S03 construimos una RACI para procesos genéricos. Ahora la vas a adaptar a procesos específicos de DataMart España.

### Tu tarea

La plantilla tiene **3 procesos pre-rellenados** y **2 vacíos**. Debes:
1. Completar los roles RACI para los procesos 4 y 5.
2. Cada fila debe tener **exactamente un "R"** y **exactamente un "A"**.

```python
# PLANTILLA RACI
# Roles: CDO, Data Owner (Negocio), Data Steward, Data Custodian (IT), DPO
# ✏️ Completa las filas de los procesos 4 y 5

raci_data = {
    "Proceso": [
        "Definición de estándares de calidad",
        "Gestión de accesos a datos sensibles",
        "Actualización del catálogo de productos",
        "...",     # ✏️ Nombre del proceso 4
        "..."      # ✏️ Nombre del proceso 5
    ],
    "CDO": ["A", "C", "I", "...", "..."],                  # ✏️
    "Data Owner": ["C", "A", "R", "...", "..."],            # ✏️
    "Data Steward": ["R", "I", "A", "...", "..."],          # ✏️
    "Data Custodian (IT)": ["I", "R", "C", "...", "..."],   # ✏️
    "DPO": ["C", "A", "I", "...", "..."]                    # ✏️
}

raci = pd.DataFrame(raci_data)
raci.set_index("Proceso", inplace=True)

# Visualización con estilo de colores
def colorear_raci(val):
    colores = {"R": "background-color: #4CAF50; color: white",
               "A": "background-color: #2196F3; color: white",
               "C": "background-color: #FFC107; color: black",
               "I": "background-color: #9E9E9E; color: white"}
    return colores.get(val, "")

raci.style.map(colorear_raci)
```

```python
# VALIDACIÓN RACI

roles = ["CDO", "Data Owner", "Data Steward", "Data Custodian (IT)", "DPO"]
errores = []

for proceso in raci.index:
    valores = raci.loc[proceso, roles].tolist()
    if "..." in valores:
        errores.append(f"  ⚠️ '{proceso}' tiene campos sin completar")
        continue
    n_r = valores.count("R")
    n_a = valores.count("A")
    if n_r != 1:
        errores.append(f"  ❌ '{proceso}' tiene {n_r} 'R' (debe tener exactamente 1)")
    if n_a != 1:
        errores.append(f"  ❌ '{proceso}' tiene {n_a} 'A' (debe tener exactamente 1)")

if errores:
    print("Problemas encontrados:")
    print("\n".join(errores))
else:
    print("✅ Matriz RACI válida: cada proceso tiene exactamente un R y un A.")
```

## Fase 4 — Plan de Calidad de Datos (10 min)

En la S04 estudiamos las **dimensiones de calidad de datos** y su importancia para la toma de decisiones. Un plan de calidad define qué dimensiones se van a medir, sobre qué datos, con qué umbrales y con qué frecuencia.

Vamos a trabajar con un dataset sintético de clientes de DataMart España y a aplicar tres dimensiones de calidad: **completitud**, **exactitud** y **consistencia**.

### Tu tarea

1. Observa el dataset sintético generado.
2. **Define los umbrales mínimos aceptables** para cada dimensión (en la celda indicada).
3. Ejecuta los checks de calidad y **escribe tu interpretación** en la celda markdown posterior.

```python
# DATASET SINTÉTICO DE CLIENTES
np.random.seed(42)
n = 200

clientes = pd.DataFrame({
    "cliente_id": range(1, n + 1),
    "nombre": [f"Cliente_{i}" for i in range(1, n + 1)],
    "email": [f"cliente{i}@email.com" if np.random.random() > 0.08
              else None for i in range(1, n + 1)],
    "telefono": [f"6{np.random.randint(10000000, 99999999)}" if np.random.random() > 0.12
                 else None for i in range(n)],
    "provincia": np.random.choice(
        ["Madrid", "Barcelona", "Valencia", "Sevilla", "madrid", "BARCELONA", None],
        size=n, p=[0.25, 0.20, 0.15, 0.10, 0.08, 0.07, 0.15]
    ),
    "fecha_registro": pd.date_range("2022-01-01", periods=n, freq="2D"),
    "gasto_total": np.round(np.random.exponential(150, n), 2)
})

# Introducir algunos duplicados
duplicados = clientes.sample(10, random_state=7).copy()
duplicados["cliente_id"] = range(n + 1, n + 11)
clientes = pd.concat([clientes, duplicados], ignore_index=True)

print(f"Dataset: {clientes.shape[0]} filas × {clientes.shape[1]} columnas")
clientes.head(10)
```

```python
# ✏️ DEFINE TUS UMBRALES DE CALIDAD
# Sustituye los valores 0 por el porcentaje mínimo aceptable (entre 0 y 100)

umbrales = {
    "completitud": 80,     # ✏️ ¿Qué % mínimo de campos NO nulos exiges? (ejemplo: 90)
    "exactitud": 90,        # ✏️ ¿Qué % mínimo de provincias correctamente formateadas? (ejemplo: 85)
    "consistencia": 100      # ✏️ ¿Qué % mínimo de registros NO duplicados? (ejemplo: 95)
}

# Verificación rápida
if any(v == 0 for v in umbrales.values()):
    print("⚠️ Define todos los umbrales antes de continuar (valores > 0)")
else:
    print(f"✅ Umbrales definidos: {umbrales}")
```

```python
# CHECKS DE CALIDAD AUTOMÁTICOS

resultados_calidad = {}

# 1. Completitud: % de valores no nulos en columnas clave
cols_clave = ["email", "telefono", "provincia"]
completitud = clientes[cols_clave].notna().mean().mean() * 100
resultados_calidad["completitud"] = round(completitud, 1)

# 2. Exactitud: % de provincias con formato correcto (primera letra mayúscula, resto minúsculas)
provincias_validas = ["Madrid", "Barcelona", "Valencia", "Sevilla"]
exactitud = clientes["provincia"].dropna().apply(lambda x: x in provincias_validas).mean() * 100
resultados_calidad["exactitud"] = round(exactitud, 1)

# 3. Consistencia: % de registros no duplicados (basado en nombre + email)
duplicados_mask = clientes.duplicated(subset=["nombre", "email"], keep="first")
consistencia = (1 - duplicados_mask.mean()) * 100
resultados_calidad["consistencia"] = round(consistencia, 1)

# Resumen
resumen = pd.DataFrame({
    "Dimensión": list(resultados_calidad.keys()),
    "Valor medido (%)": list(resultados_calidad.values()),
    "Umbral mínimo (%)": [umbrales.get(k, 0) for k in resultados_calidad.keys()]
})
resumen["¿Cumple?"] = resumen.apply(
    lambda row: "✅ Sí" if row["Valor medido (%)"] >= row["Umbral mínimo (%)"] else "❌ No",
    axis=1
)

print("=== INFORME DE CALIDAD DE DATOS ===\n")
print(resumen.to_string(index=False))
```

## Fase 6 — Resumen Ejecutivo y Radar de Madurez (8 min)

El radar de madurez ofrece una visión rápida del estado del plan de gobierno. Cada eje representa una de las cinco áreas que has trabajado y se puntúa de 0 a 5 según el nivel de desarrollo alcanzado en tu plan.

```python
# ✏️ PUNTUACIÓN DE MADUREZ (1 = incipiente, 5 = maduro)
# Valora el nivel de desarrollo de cada área EN TU PLAN

madurez = {
    "Estrategia\n(Objetivos)": 4,     # ✏️ Puntúa de 1 a 5
    "Roles y\nResponsabilidades": 3,   # ✏️
    "Políticas": 3,                     # ✏️
    "Calidad\nde Datos": 1,            # ✏️
    "Privacidad\ny RGPD": 3            # ✏️
}

if any(v == 0 for v in madurez.values()):
    print("⚠️ Asigna una puntuación (1-5) a cada dimensión antes de generar el radar.")
else:
    # Generar radar
    categorias = list(madurez.keys())
    valores = list(madurez.values())
    N = len(categorias)

    angulos = [n / float(N) * 2 * np.pi for n in range(N)]
    valores_plot = valores + [valores[0]]
    angulos_plot = angulos + [angulos[0]]

    fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))
    ax.plot(angulos_plot, valores_plot, 'o-', linewidth=2, color='#2196F3')
    ax.fill(angulos_plot, valores_plot, alpha=0.25, color='#2196F3')
    ax.set_xticks(angulos)
    ax.set_xticklabels(categorias, size=10)
    ax.set_ylim(0, 5)
    ax.set_yticks([1, 2, 3, 4, 5])
    ax.set_yticklabels(["1", "2", "3", "4", "5"], size=8)
    ax.set_title("Radar de Madurez — Plan de Gobierno del Dato\nDataMart España", size=13, pad=20)
    plt.tight_layout()
    plt.show()

    media = np.mean(valores)
    print(f"\nPuntuación media de madurez: {media:.1f} / 5.0")
    if media < 2:
        print("📊 Nivel: Inicial — el plan necesita más desarrollo.")
    elif media < 3.5:
        print("📊 Nivel: En desarrollo — buen punto de partida, pero hay áreas de mejora.")
    else:
        print("📊 Nivel: Avanzado — el plan tiene una base sólida.")
```
