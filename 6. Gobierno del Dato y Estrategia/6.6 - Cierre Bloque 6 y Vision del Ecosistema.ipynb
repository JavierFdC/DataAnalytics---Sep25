## Parte 1 — Resumen del Bloque 6

### El hilo conductor: tres preguntas que ninguna organización puede ignorar

Si tuviéramos que resumir el Bloque 6 en tres preguntas, serían estas:

| Pregunta clave | Sesiones que la responden | Caso de estudio |
|---|---|---|
| **¿Cómo organizamos los datos?** | S01: Arquitectura de datos | Hospital Santa Clara |
| **¿Quién es responsable de qué?** | S02: Data Hub · S03: Marco RACI + Políticas | Hospital SC → DataMart España |
| **¿Cómo garantizamos que los datos son fiables y seguros?** | S04: Calidad + Metadatos + RGPD | DataMart España |
| **¿Cómo se integra todo en un plan real?** | S05: Reto Plan de Gobierno | DataMart España |

Estas preguntas no son independientes, sino que se construyen unas sobre otras:

- Sin una arquitectura bien pensada, el gobierno no tiene sobre qué aplicarse. Si no sabemos dónde viven los datos ni cómo fluyen, no podemos asignar responsabilidades ni medir calidad.
- Sin roles claros (Data Owner, Data Steward, DPO), la calidad no tiene responsable. Cuando nadie es responsable de un dato, nadie lo cuida.
- Sin calidad garantizada, los dashboards de Power BI mostrarán cifras incorrectas, los modelos de IA aprenderán sobre datos sesgados y las decisiones de negocio serán erróneas.
- Sin privacidad, la organización incumple el RGPD y se expone a sanciones de hasta el 4% de su facturación global anual.

El Bloque 6 no es un bloque opcional "de gestión" que se añade al final. Es el marco que hace que todo lo demás funcione correctamente.

### Por qué el patrón Bronze-Silver-Gold se ha convertido en estándar

Antes de ver la visualización del flujo completo, vale la pena entender de dónde viene este patrón y por qué ha ganado tanto peso en la industria.

El patrón —también llamado **arquitectura Medallion**— lo popularizó Databricks alrededor de 2019 en el contexto de Delta Lake. La idea es sencilla pero poderosa: en lugar de transformar los datos una sola vez y perder el original, se guardan tres versiones de cada dato en capas de calidad creciente.

**¿Por qué es tan valioso?**

- **Recuperabilidad**: si una transformación fallida corrompe los datos de Silver o Gold, siempre puedes reprocesar desde Bronze. No hay pérdida de datos irrecuperable.
- **Trazabilidad**: cualquier analista puede ver exactamente qué dato llegó de la fuente (Bronze), cómo quedó tras limpiarlo (Silver) y qué métricas se calcularon sobre él (Gold). Esto es linaje de datos en su forma más concreta.
- **Separación de responsabilidades**: el equipo de ingeniería gestiona Bronze (garantiza que los datos llegan). El equipo de governance gestiona Silver (garantiza la calidad). El equipo de análisis gestiona Gold (garantiza que las métricas son correctas para el negocio).
- **Compatibilidad con el RGPD**: los datos personales se pueden seudonimizar o anonimizar en el paso de Bronze a Silver, garantizando que la capa Gold —que es la que usan más personas— nunca exponga datos personales crudos.

El patrón es agnóstico a la tecnología: se puede implementar en AWS, Azure, Google Cloud, Databricks o incluso en un servidor local. Lo que importa es el principio, no la herramienta.

````python
# El flujo que conecta arquitectura, gobierno y consumo
fig, ax = plt.subplots(figsize=(13, 3.8))
ax.set_xlim(0, 13)
ax.set_ylim(0, 3.8)
ax.axis('off')
etapas = [
    ("BRONZE\nDatos crudos", "#cd7f32", "S01", "Copia fiel\ndel origen"),
    ("SILVER\nDatos limpios", "#7f8c8d", "S01 + S04", "Calidad\naplicada"),
    ("GOLD\nDatos listos", "#f1c40f", "S01 + S04", "Listos para\nconsumidores"),
    ("CONSUMO\nGobernado", "#2ecc71", "S02 - S05", "Roles · Accesos\nPrivacidad"),
    ("DECISIÓN\nde Negocio", "#2980b9", "Bloques 7-8", "Dashboards\ny KPIs"),
]
x_pos = [1.1, 3.5, 5.9, 8.3, 10.8]
for i, (nombre, color, sesion, desc) in enumerate(etapas):
    x = x_pos[i]
    rect = mpatches.FancyBboxPatch((x-0.85, 0.9), 1.7, 1.6,
                                    boxstyle="round,pad=0.1", fc=color, ec='white', alpha=0.88, lw=2)
    ax.add_patch(rect)
    tc = '#2c3e50' if color == '#f1c40f' else 'white'
    ax.text(x, 1.9, nombre, ha='center', va='center', fontsize=8, fontweight='bold', color=tc)
    ax.text(x, 0.62, desc, ha='center', va='center', fontsize=7, color='#555')
    ax.text(x, 2.72, sesion, ha='center', va='center', fontsize=7, color=color, fontweight='bold',
            bbox=dict(boxstyle='round,pad=0.2', fc='white', ec=color, lw=1.2))
    if i < 4:
        ax.annotate('', xy=(x_pos[i+1]-0.92, 1.7), xytext=(x+0.92, 1.7),
                    arrowprops=dict(arrowstyle='->', color='#95a5a6', lw=2))
gov_rect = mpatches.FancyBboxPatch((0.2, 0.15), 9.9, 3.35,
                                    boxstyle="round,pad=0.1", fc='none', ec='#e74c3c', lw=2.2, linestyle='--')
ax.add_patch(gov_rect)
ax.text(5.17, 3.63, "← Gobernado por el Bloque 6: Arquitectura · Roles · Calidad · RGPD →",
        ha='center', va='center', fontsize=8, color='#e74c3c', fontweight='bold')
plt.title("Del dato crudo a la decisión de negocio: el flujo gobernado", fontsize=11, pad=5)
plt.tight_layout()
plt.show()
````

## Parte 2 — El Ecosistema Completo del Curso

### ¿Por qué parece todo desconectado?

Esta es la pregunta que más se repite al avanzar en el curso, y tiene una respuesta clara.

**El motivo real:** el curso está estructurado de forma pedagógica (primero aprendes SQL, luego Python, luego Cloud...), pero la realidad profesional no está organizada así. En una empresa real, SQL, Python, Cloud y Power BI no son etapas que suceden una después de otra: son **capas que coexisten simultáneamente** sobre los mismos datos, en el mismo momento, gestionadas por equipos distintos.

Cuando estudias SQL de forma aislada, aprendes a escribir consultas. Pero no ves que esas consultas se ejecutan sobre bases de datos que un Data Engineer diseñó (Python/Cloud), cuyos datos han sido validados por un equipo de governance (Bloque 6), y cuyos resultados alimentarán un dashboard de Power BI que verá el Director Comercial mañana por la mañana.

**El cambio de modelo mental que necesitas:**

No es esto (cadena lineal, cómo se enseña):
```
[SQL] → [Python] → [Cloud] → [Gobierno] → [Power BI]
```

Es esto (stack por capas, cómo funciona en la realidad):
```
        ┌─────────────────────────────────────────────────────┐
        │  GOBIERNO DEL DATO — aplica a TODAS las capas       │
        └─────────────────────────────────────────────────────┘
              ↕          ↕           ↕             ↕
  [FUENTES] [ALMACENAMIENTO] [PROCESAMIENTO] [VISUALIZACIÓN]
  (Bloque 1)  (Bloques 1,5)    (Bloques 2-5)   (Bloques 7-8)
```

La diferencia es fundamental: en el modelo de cadena, el Bloque 6 sería "el paso de gobierno". En el modelo de stack, el Bloque 6 es el **marco transversal** que gobierna todos los demás pasos simultáneamente. No es lo mismo diseñar una pipeline Python sin pensar en calidad de datos que diseñarla con las reglas de validación del Bloque 6 ya incorporadas.

**La visualización que viene a continuación** muestra exactamente este stack. Tómate un momento para leerla de abajo a arriba: cada capa añade valor sobre la anterior, y la columna roja de la izquierda (Gobierno del Dato) aplica a todas.

### El viaje de un dato — de la tienda a la decisión

Una forma muy efectiva de consolidar el modelo mental del stack es seguir un único dato concreto a lo largo de todo su recorrido. Esto es lo que hacen los arquitectos de datos cuando diseñan un nuevo sistema: rastrean mentalmente un dato representativo y se preguntan "¿qué le pasa en cada paso?"

El dato que vamos a seguir es una **compra realizada por un cliente en una tienda de DataMart España**. Es un dato sencillo (un importe, una fecha, un artículo), pero para que ese dato acabe en un dashboard de ventas que el director regional consulta cada mañana, tiene que atravesar todos los bloques del curso.

Observa en el diagrama siguiente cómo el Bloque 6 no ocupa un único nodo: **aparece en tres pasos consecutivos** (Bronze, Silver y Gold), porque el gobierno del dato no es un momento puntual, sino una disciplina que aplica durante toda la transformación del dato. En Bronze garantiza que la copia cruda sea fiel al origen. En Silver garantiza que la limpieza se hace según las reglas acordadas. En Gold garantiza que las métricas están definidas de forma unívoca y que el acceso está controlado.

````python
# El viaje de una compra a través de todo el ecosistema
fig, ax = plt.subplots(figsize=(15, 5.5))
ax.set_xlim(0, 15)
ax.set_ylim(0, 5.5)
ax.axis('off')
pasos = [
    ("Compra\nen tienda", "#95a5a6", "Fuente\n(TPV)", "Dato bruto\norigen"),
    ("Base de\ndatos SQL", "#2980b9", "Bloque 1", "INSERT INTO\ntransacciones"),
    ("Pipeline\nETL Python", "#8e44ad", "Bloques 2-4", "pd.read_sql()\ntransform()"),
    ("Data Lake\nBronze", "#cd7f32", "Bloque 6\nArquitectura", "Copia cruda\nsin modificar"),
    ("Limpieza\nSilver", "#7f8c8d", "Bloque 6\nCalidad", "Validación de\n6 dimensiones"),
    ("KPIs Gold\nGobernado", "#f1c40f", "Bloque 6\nGobierno", "Roles · Linaje\nPrivacidad"),
    ("Dashboard\nPower BI", "#e67e22", "Bloque 7", "DAX · Modelo\nestrella"),
    ("Decisión\ndel negocio", "#27ae60", "Negocio", "Abrir nueva\ntienda"),
]
x_pos = [0.8, 2.7, 4.6, 6.5, 8.4, 10.3, 12.2, 14.1]
y_c = 3.1
for i, (titulo, color, bloque, ejemplo) in enumerate(pasos):
    x = x_pos[i]
    rect = mpatches.FancyBboxPatch((x-0.75, y_c-0.75), 1.5, 1.5,
                                    boxstyle="round,pad=0.1", fc=color, ec='white', alpha=0.9, lw=1.8)
    ax.add_patch(rect)
    tc = '#2c3e50' if color == '#f1c40f' else 'white'
    ax.text(x, y_c+0.2, titulo, ha='center', va='center', fontsize=7.5, fontweight='bold', color=tc)
    ax.text(x, y_c-0.4, bloque, ha='center', va='center', fontsize=6.5, color=tc, alpha=0.88)
    ax.text(x, y_c-1.3, ejemplo, ha='center', va='center', fontsize=6.5, color='#444', style='italic')
    if i < 7:
        ax.annotate('', xy=(x_pos[i+1]-0.8, y_c), xytext=(x+0.8, y_c),
                    arrowprops=dict(arrowstyle='->', color='#7f8c8d', lw=2))
brace_rect = mpatches.FancyBboxPatch((5.7, 1.25), 5.35, 0.55,
                                      boxstyle="round,pad=0.06", fc='#e74c3c', ec='#c0392b', alpha=0.82)
ax.add_patch(brace_rect)
ax.text(8.37, 1.52, "GOBIERNO DEL DATO (Bloque 6) — atraviesa Bronze, Silver y Gold",
        ha='center', va='center', fontsize=7.5, color='white', fontweight='bold')
ax.text(7.5, 0.55,
        "Clave: el Bloque 6 no es un paso más en la cadena, ATRAVIESA las capas de almacenamiento y procesamiento",
        ha='center', va='center', fontsize=8, color='#e74c3c', style='italic')
plt.title("El viaje de un dato: desde una compra en tienda hasta la decisión de negocio",
          fontsize=11, fontweight='bold', pad=8)
plt.tight_layout()
plt.show()
````

### Los roles del mercado y qué bloques los preparan

Uno de los aspectos que más preocupa al terminar el curso es saber para qué rol profesional te estás preparando. La respuesta no es única: el curso cubre todas las capas del ecosistema, por lo que abre puertas a varios perfiles.

Lo que diferencia a un Data Analyst de un Data Engineer no es solo qué herramientas usa, sino **a qué capa del stack dedica más tiempo** y **qué tipo de preguntas responde**:

- Un **Data Analyst** vive principalmente en las capas de analítica y visualización. Su materia prima son los datos limpios de la capa Gold. Su producto son los informes y dashboards que informan las decisiones.
- Un **Data Engineer** vive en las capas de almacenamiento y procesamiento. Construye los pipelines que mueven los datos desde las fuentes hasta la capa Gold que consumirá el analista.
- Un **Data Scientist** trabaja sobre la capa de analítica con técnicas estadísticas y de machine learning. Necesita datos de calidad (Bloque 6) porque un modelo entrenado sobre datos con sesgos o inconsistencias producirá predicciones incorrectas.
- Un **Data Governance Analyst** es transversal: su trabajo es garantizar que todas las capas funcionan con datos fiables, seguros y conformes. Interacciona con todos los demás perfiles.
- Un **Analytics Engineer** —perfil emergente muy demandado— es el puente entre el Data Engineer y el Data Analyst. Transforma los datos crudos en métricas de negocio bien documentadas usando herramientas como dbt.

El mapa de calor que sigue muestra de un vistazo qué bloques son críticos para cada rol. Úsalo para orientar tus próximos pasos de aprendizaje.

### El Bloque 6 como el "sistema operativo" de los datos

Hay una metáfora que suele hacer que todo encaje de golpe: si las herramientas del resto del curso son las **aplicaciones** (SQL es una aplicación de consulta de bases de datos, Python es una aplicación de procesamiento, Power BI es una aplicación de visualización), entonces el gobierno del dato es el **sistema operativo** sobre el que todas ellas funcionan.

Sin sistema operativo, las aplicaciones no saben cómo acceder al disco, no hay control de quién puede ejecutar qué, y si dos procesos intentan escribir en el mismo fichero simultáneamente, los datos se corrompen. Sin gobierno del dato, ocurre exactamente lo mismo:

| Sin Gobierno del Dato | Con Gobierno del Dato |
|---|---|
| "Cada departamento tiene su Excel con cifras distintas" | Fuente única de verdad con Data Hub (MDM) |
| "Los datos de facturación no cuadran con los del CRM" | Integración centralizada y bidireccional |
| "No sé si el campo `monto` incluye o no el IVA" | Diccionario de datos con definición acordada |
| "No sabemos quién puede acceder a los datos de clientes" | Política de acceso + Matriz RACI |
| "Tuvimos una brecha pero no lo notificamos a tiempo" | Protocolo de notificación AEPD en 72 horas |
| "Tenemos petabytes de datos pero no sabemos si son fiables" | Scorecard de calidad con umbrales y SLAs |

Lo más importante: el gobierno del dato **no ralentiza el trabajo** de los analistas, ingenieros y científicos de datos. Al contrario, lo acelera. Cuando sabes exactamente qué significa cada campo, quién lo produce, cómo se validó y si cumple los umbrales de calidad, puedes empezar a analizarlo de inmediato sin perder horas investigando si el dato es fiable.

### ARQUITECTURA DE DATOS

**Bronze (Capa Bronze)** — Primera capa de la arquitectura Medallion. Almacena los datos tal como llegan de la fuente, sin ninguna transformación: con todos sus errores, duplicados y formatos heterogéneos. Es la "caja fuerte" del dato crudo. Permite reprocesar datos desde el origen si una transformación posterior falla.

**Silver (Capa Silver)** — Segunda capa de la arquitectura Medallion. Contiene datos limpios, validados y estandarizados: tipos corregidos, duplicados eliminados, nulos tratados, reglas de negocio básicas aplicadas. Es la "verdad limpia" de la organización, la base sobre la que se construyen los análisis.

**Gold (Capa Gold)** — Tercera capa de la arquitectura Medallion. Contiene datos agregados, enriquecidos y listos para consumo directo: métricas precalculadas, tablas dimensionales para dashboards, datasets para modelos de ML. Es lo que consume el usuario final y lo que alimenta Power BI o Tableau.

**Data Warehouse (Almacén de datos)** — Sistema centralizado para almacenar datos estructurados, optimizado para consultas analíticas. Los datos se transforman antes de cargarse (schema-on-write), lo que garantiza consistencia pero añade rigidez. Ideal para reporting y BI tradicional. Ejemplos: Snowflake, BigQuery, Amazon Redshift, Azure Synapse.

**Data Lake (Lago de datos)** — Repositorio que almacena datos en su formato original —estructurados, semiestructurados o no estructurados— sin transformación previa. El esquema se aplica al consultar (schema-on-read). Muy flexible y económico, pero sin gobernanza puede convertirse en un "data swamp". Ejemplos: AWS S3, Azure Data Lake Storage, Google Cloud Storage.

**Data Lakehouse** — Arquitectura híbrida que combina la flexibilidad del Data Lake con las garantías transaccionales del Data Warehouse (ACID, versionado, esquemas). Tecnologías como Delta Lake, Apache Iceberg o Apache Hudi añaden estas capacidades sobre ficheros en el lake. Es la tendencia dominante en arquitecturas modernas.

**Data Mesh** — Enfoque descentralizado donde cada dominio de negocio es responsable de sus propios datos como si fueran un producto ("data product"). No es una tecnología, es un modelo organizativo. Requiere alta madurez técnica y cultural. Escala mejor que los enfoques centralizados en organizaciones grandes y complejas.

**Data Swamp ("pantano de datos")** — Estado en que cae un Data Lake sin gobierno: los datos se acumulan sin catalogación ni documentación, los esquemas son incoherentes y nadie sabe qué hay ni cómo usarlo. Principal riesgo de un Data Lake sin programa de gobierno del dato.

**ETL (Extract, Transform, Load)** — Proceso que extrae datos de fuentes, los transforma (limpia, normaliza, enriquece) y los carga en el destino. La transformación ocurre ANTES de cargar. Patrón clásico para alimentar Data Warehouses. La transformación se realiza en un servidor externo.

**ELT (Extract, Load, Transform)** — Variante donde los datos se cargan primero en su formato crudo y la transformación ocurre DESPUÉS, aprovechando la potencia de cómputo del propio data lake o warehouse. Más habitual en arquitecturas modernas con grandes volúmenes. dbt es la herramienta ELT por excelencia.

**OLAP (Online Analytical Processing)** — Sistemas optimizados para consultas analíticas complejas sobre datos históricos: muchas lecturas, pocas escrituras. Son los Data Warehouses y herramientas BI. Los cubos OLAP permiten consultas multidimensionales muy rápidas.

**OLTP (Online Transaction Processing)** — Sistemas operacionales optimizados para muchas transacciones pequeñas en tiempo real: inserciones, actualizaciones, lecturas rápidas. Son las bases de datos de los sistemas de negocio (CRM, ERP, POS). Son las fuentes de datos del pipeline analítico.

**Schema-on-write** — El esquema (estructura, tipos) se define y aplica al escribir los datos. Garantiza consistencia pero requiere conocer la estructura de antemano. Patrón del Data Warehouse. Si el esquema cambia, hay que hacer una migración.

**Schema-on-read** — El esquema se aplica al consultar los datos. Los datos se almacenan "tal cual" y cada consumidor los interpreta con el esquema que necesite. Patrón del Data Lake. Muy flexible pero puede generar inconsistencias entre consumidores.

### GOBIERNO DEL DATO

**CDO (Chief Data Officer)** — Director/a de Datos. Ejecutivo/a responsable de la estrategia de datos de la organización: define la visión, prioriza inversiones y alinea el programa de gobierno con los objetivos de negocio. Es el "A" (Accountable) de las decisiones estratégicas sobre datos.

**Comité de Datos (Data Governance Council)** — Órgano colegiado que establece la estrategia, aprueba políticas transversales y resuelve conflictos entre dominios. Incluye representantes de negocio, TI, legal y dirección. Es el máximo órgano de gobierno del dato en la organización.

**DAMA-DMBOK** — (Data Management Body of Knowledge) Marco de referencia publicado por DAMA International que define 11 áreas de gestión de datos (calidad, metadatos, seguridad, arquitectura, etc.) y el vocabulario estándar de la industria. Es el "manual de instrucciones" del dato a nivel global.

**Data Governance (Gobierno del dato)** — Conjunto de políticas, procesos, roles y métricas que aseguran que los datos de una organización son correctos, seguros, accesibles y útiles. No es un proyecto con fecha de fin: es un programa continuo. Abarca calidad, seguridad, ciclo de vida, cumplimiento legal y generación de valor.

**Data Hub** — Nodo central de integración bidireccional que actúa como "notario" de los datos maestros. A diferencia del Data Warehouse (unidireccional), el Data Hub sincroniza datos en todas las direcciones: si un sistema actualiza un dato, el Hub lo propaga al resto. Implementa MDM.

**Data Owner (Propietario del dato)** — Responsable de negocio que "posee" un dominio de datos. Decide quién accede, aprueba definiciones y es el responsable último de la calidad. Rol organizativo, no técnico. Ej: el Director Comercial es Data Owner de los datos de clientes.

**Data Steward (Administrador del dato)** — Perfil operativo que implementa las políticas del Data Owner en el día a día: monitoriza calidad, gestiona incidencias, documenta metadatos. Es el "guardián cotidiano" del dato. Trabaja en el negocio, no en TI.

**Data Custodian (Custodio técnico)** — Perfil de TI responsable de la infraestructura técnica: almacenamiento, backups, seguridad de acceso, rendimiento. No decide QUÉ datos se guardan, sino CÓMO se guardan de forma segura y eficiente.

**DPO (Data Protection Officer / Delegado de Protección de Datos)** — Figura exigida por el RGPD en organizaciones que tratan datos sensibles a gran escala. Supervisa el cumplimiento normativo, asesora sobre evaluaciones de impacto y actúa como enlace con la AEPD. Independiente de la dirección de la empresa.

**Marco de madurez** — Escala 1-5 (Inicial → Repetible → Definido → Gestionado → Optimizado) que evalúa el nivel de un programa de gobierno en dimensiones como Personas, Procesos, Tecnología, Datos y Cultura. Permite un roadmap de mejora realista.

**Master Data Management (MDM)** — Disciplina que garantiza una única versión de referencia para las entidades críticas de negocio (cliente, producto, proveedor, empleado). Evita que el mismo cliente aparezca con tres nombres distintos en tres sistemas diferentes.

**Matriz RACI** — Herramienta que asigna responsabilidades por actividad: R (Responsible/ejecuta), A (Accountable/rinde cuentas — único por actividad), C (Consulted/opina antes de decidir), I (Informed/recibe notificación al finalizar). Resuelve el problema de "nadie sabe quién hace qué".

**Política de datos** — Documento formal que establece reglas, estándares y procedimientos para gestionar un aspecto concreto de los datos (acceso, calidad, retención, privacidad). Tiene un propietario, una frecuencia de revisión y un estado (vigente / borrador / no iniciada).

### CALIDAD, METADATOS Y LINAJE

**Catálogo de datos** — Plataforma que permite buscar, descubrir y entender los datasets disponibles en una organización. El diccionario de datos es un componente del catálogo. Ejemplos: Apache Atlas, Alation, Microsoft Purview, Collibra.

**Completitud (Completeness)** — Dimensión de calidad: qué porcentaje de los valores que deberían estar presentes realmente lo están. Ej: el 92% de los registros de cliente tienen email → completitud del 92%.

**Consistencia (Consistency)** — Dimensión de calidad: si el mismo dato tiene el mismo valor en todos los sistemas. Ej: ¿el nombre del cliente es igual en el CRM y en facturación? La falta de consistencia es la fuente más frecuente de "informes contradictorios".

**Diccionario de datos** — Documento que describe campo por campo cada tabla: nombre, tipo, descripción de negocio, propietario, reglas de validación, ejemplo. Es el contrato entre quien produce el dato y quien lo consume.

**Exactitud (Accuracy)** — Dimensión de calidad: si los valores reflejan fielmente la realidad. Ej: ¿la dirección del cliente es la real? Difícil de medir sin cruzar con una fuente de referencia externa. La más difícil de medir de las seis dimensiones.

**Linaje de datos (Data Lineage)** — Documentación del origen, transformaciones y destino de cada dato: de dónde viene, qué pipelines lo han procesado, qué informes lo consumen. Permite detectar errores y es obligatorio bajo el RGPD (art. 30: registro de actividades de tratamiento).

**Metadatos** — "Datos sobre los datos". Tipos: técnicos (tipos, esquemas, restricciones), de negocio (definiciones, propietarios, reglas) y operacionales (fecha de actualización, volumen, pipeline que lo alimenta). Sin metadatos, un campo llamado `mnt_bruto` es un misterio.

**Oportunidad (Timeliness)** — Dimensión de calidad: si los datos están disponibles cuando se necesitan. Ej: ¿el stock de la web se actualiza antes de que abra el comercio online? Un dato correcto pero tardío puede ser tan dañino como uno incorrecto.

**Scorecard de calidad** — Cuadro de mando que mide periódicamente el cumplimiento de las dimensiones de calidad para un dominio de datos concreto. Permite detectar degradaciones antes de que afecten a los análisis o dashboards.

**SLA de datos (Service Level Agreement)** — Acuerdo formal sobre los niveles mínimos de calidad: ej. "la tabla de clientes tendrá como máximo un 2% de emails inválidos y se actualizará antes de las 8:00 cada día laboral". Define el umbral de calidad aceptable entre el productor y el consumidor del dato.

**Unicidad (Uniqueness)** — Dimensión de calidad: ausencia de duplicados. Ej: ¿el mismo cliente aparece solo una vez en la tabla? Los duplicados inflan métricas, sesgan análisis y dificultan la visión 360 del cliente.

**Validez (Validity)** — Dimensión de calidad: si los valores cumplen las reglas de formato y dominio. Ej: ¿los emails tienen formato usuario@dominio.ext? ¿Las edades están entre 0 y 120? ¿Los códigos postales tienen exactamente 5 dígitos?

### PRIVACIDAD Y CUMPLIMIENTO LEGAL

**AEPD (Agencia Española de Protección de Datos)** — Autoridad de control española en materia de protección de datos. Recibe las notificaciones de brechas de seguridad (obligatorio en 72h si hay riesgo para los afectados) y puede imponer sanciones de hasta 20 millones de euros o el 4% de la facturación global anual.

**Anonimización** — Proceso irreversible que elimina cualquier posibilidad de identificar a la persona a partir del dato. Un dato anonimizado deja de ser dato personal a efectos del RGPD. Técnicas: supresión, generalización, perturbación, agregación. Caso de uso: datasets públicos, investigación abierta.

**Base legal** — Fundamento jurídico que legitima el tratamiento de datos personales bajo el RGPD. Las más comunes: consentimiento explícito (libre, informado, específico), ejecución de un contrato, obligación legal e interés legítimo. Sin base legal, el tratamiento es ilícito.

**Derechos ARCO** — Acceso, Rectificación, Cancelación/Supresión y Oposición. Derechos que el RGPD reconoce a los interesados sobre sus datos personales. Las organizaciones deben responder en 30 días. El DPO coordina su tramitación. Desde el RGPD se añaden también portabilidad y limitación.

**EIPD (Evaluación de Impacto en Protección de Datos)** — Análisis obligatorio antes de iniciar tratamientos de alto riesgo (perfilado masivo, videovigilancia, datos sensibles a gran escala). Identifica riesgos y medidas de mitigación. También conocida como DPIA (Data Protection Impact Assessment).

**LOPD-GDD** — Ley Orgánica de Protección de Datos y Garantía de los Derechos Digitales. Norma española que adapta el RGPD al ordenamiento jurídico nacional, añadiendo aspectos como el derecho al olvido digital y la protección frente al acoso en redes sociales.

**PII (Personally Identifiable Information)** — Datos que identifican o pueden identificar a una persona física: nombre, email, teléfono, DNI, dirección IP, datos de localización, número de cookie, etc. Requieren las protecciones más estrictas bajo el RGPD. También existen datos de categorías especiales (salud, ideología, origen étnico) con protección reforzada.

**RGPD / GDPR** — Reglamento General de Protección de Datos (General Data Protection Regulation). Reglamento europeo en vigor desde mayo de 2018. Define principios (minimización, limitación de finalidad, integridad), derechos (ARCO + portabilidad + limitación), obligaciones (DPO, registro de actividades, notificación de brechas en 72h) y sanciones.

**Seudonimización** — Proceso reversible que reemplaza identificadores directos por un seudónimo (generalmente un hash con sal). El dato sigue siendo personal porque puede revertirse con la clave. El RGPD la recomienda como medida de seguridad (art. 32). Caso de uso: analítica interna, entornos de desarrollo y test.

### ROLES, HERRAMIENTAS Y OTROS TÉRMINOS

**Analytics Engineer** — Perfil híbrido entre Data Engineer y Data Analyst: transforma datos crudos en métricas de negocio fiables y bien documentadas usando herramientas como dbt. Garantiza que los datos que llegan a Power BI sean correctos. Perfil muy demandado desde 2020.

**Apache Airflow** — Plataforma open-source para orquestar pipelines de datos mediante DAGs (Directed Acyclic Graphs). Permite programar, monitorizar y gestionar flujos ETL/ELT complejos con dependencias entre tareas.

**dbt (data build tool)** — Herramienta open-source que permite transformar datos en el warehouse usando SQL con buenas prácticas de ingeniería (tests, documentación, versionado, linaje automático). Es la herramienta central del perfil Analytics Engineer e implementa el patrón ELT.

**Delta Lake** — Capa de almacenamiento open-source creada por Databricks que añade transacciones ACID, versionado (time travel), esquemas y optimización de lecturas sobre ficheros Parquet en un Data Lake. Es la base tecnológica de la arquitectura Lakehouse.

**EDA (Exploratory Data Analysis / Análisis Exploratorio de Datos)** — Primera fase de cualquier proyecto de análisis o ciencia de datos: explorar la distribución de los datos, detectar outliers, analizar correlaciones y entender la estructura antes de construir modelos o dashboards.

**KPI (Key Performance Indicator)** — Indicador Clave de Rendimiento. Métrica cuantificable que evalúa el progreso hacia un objetivo de negocio. Una KPI mal calculada por datos de baja calidad lleva a decisiones incorrectas. El gobierno del dato garantiza que las KPIs tienen una definición única y datos fiables.

**Microsoft Purview (Azure Purview)** — Plataforma de gobierno del dato de Microsoft integrada en Azure. Proporciona catálogo de datos, linaje automático, clasificación de datos sensibles (PII) y cumplimiento RGPD. Es la herramienta de gobierno más usada en entornos de la nube de Microsoft.

**Modelo estrella (Star Schema)** — Modelo de datos para analítica donde una tabla de hechos (ventas, pedidos, eventos) está rodeada de tablas de dimensiones (cliente, producto, tiempo, tienda). Es el modelo de referencia en Power BI y herramientas de BI. Maximiza el rendimiento de las consultas analíticas.

**Pipeline de datos** — Conjunto de procesos automatizados que mueven y transforman datos desde las fuentes hasta el destino final. Incluye extracción, transformación, validación y carga. Puede ser batch (por lotes, ej: cada noche) o streaming (en tiempo real, ej: cada mensaje del IoT).

**RLS (Row-Level Security)** — Seguridad a nivel de fila. Mecanismo que restringe qué filas de datos puede ver cada usuario en función de su rol.
